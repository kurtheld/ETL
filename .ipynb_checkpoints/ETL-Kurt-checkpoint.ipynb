```python
#--create an empty list to store text of divs

our_list = []

#--Create a soup object to find all text of table data

quotes = soup.find_all('div', class_='item')

#--loop through and store text to list

for quote in quotes:
        for cell in quote.find_all('div'):
            our_list.append(cell.text)
            

#--Retrive column values for DataFrame            
columns = our_list [:3]
#columns

our_list
```

```python
# Convert list of Lists of Country GDP per Capita into a dataframe
df_17 = pd.DataFrame(quotes, columns = [columns[0] , columns[1], columns[2],"Empty"])
df_17
```

#### Normalizing Data

```python


#df_17.drop(df_17.index[0],inplace=True)

#df_17 = df_17.drop(["Empty"], axis=1)

#df_17.reset_index(inplace = True)

#df_17 = df_17.drop(["index"], axis =1)

df_17.head()
```

```python
##--cleaning <div> tag from countries_bs4 list

countries_bs4 = df_17.Country.tolist()

countries_str = list(map(str, countries_bs4))

print(countries_str[0][5:-6])

countries = []

for country in countries_str:
    if country not in countries:
        value = country[5:-6]
        countries.append(value)
countries[:5]
```

```python
#--cleaning <div> tag from 1960 column: f_year

f_year_bs4 = df_17["1960"].tolist()
f_year_bs4 = list(map(str,f_year_bs4))

f_year_bs4[0][5:-6]

f_year = []

for data in f_year_bs4:
    if data == "<div></div>":
        value = "N/A"
        f_year.append(value)
    else:
        value = data[5:-6]
        f_year.append(value)

f_year[0:5]
```

```python
#--cleaning <div> tag from 2017 column: s_year

s_year_bs4 = df_17["2017"].tolist()
s_year_bs4 = list(map(str, s_year_bs4))

s_year_bs4[0][5:-6]

s_year = []

for data in s_year_bs4:
    if data == "<div></div>":
        value = "N/A"
        s_year.append(value)
    else:
        value = data[5:-6]
        s_year.append(value)

s_year[0:5]
```

#### Final DataFrame for Year 2017

```python
#Create a prototype DataFrame from the lists created: proto_df

proto_df = pd.DataFrame({
    
            "Country": countries,
            "1960": f_year,
            "2017": s_year,
})

proto_df.head()
```
