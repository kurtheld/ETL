```python
 # Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
from splinter import Browser
import pandas as pd
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
 # URL of page to be scraped
url = "https://data.worldbank.org/indicator/NY.GDP.PCAP.CD"
browser.visit(url)
```

```python
 # Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(browser.html, 'html.parser')
```

```python
quotes = soup.find_all('div', class_='item')

for quote in quotes:
        for cell in quote.find_all('div'):
            print(cell.text)
```

```python
# Convert list of Lists of Country GDP per Capita into a dataframe
GDPPerCap17 = pd.DataFrame(quotes, columns = ['Country' , '1960', '2017',"Empty"])
GDPPerCap17['Country'] = GDPPerCap17['Country'].map(lambda x: str(x)[5:-6])
GDPPerCap17['1960'] = GDPPerCap17['1960'].map(lambda x: str(x)[5:-6])
GDPPerCap17['2017'] = GDPPerCap17['2017'].map(lambda x: str(x)[5:-6])
GDPPerCap17['Empty'] = GDPPerCap17['Empty'].map(lambda x: str(x)[5:-6])
# GDPPerCap17['Country'] = GDPPerCap17['Country'].map(lambda x: str(x)[-6:])
GDPPerCap17
```

```python
GDPPerCap17.to_csv("GDPPerCap17.csv")
```
